```python
df = pd.read_csv("../week--02/data/processed/CEHHbInToW_outliers_handled_values.csv")
df.head()
```

```python
df["Age"].min(), df["Age"].max()
```

350 - 580 => poor
580 - 670 => fair
670 - 740 => good
740 - 800 => very good
800 - 850 => excellent

```python
def custom_binning_credit_score(score):
    if score < 580:
        return "poor"
    if score < 670:
        return "fair"
    if score < 740:
        return "good"
    if score < 800:
        return "very good"
    if score <= 850:
        return "excellent"
    else:
        assert True, "Invalid credit score"
        
df["CreditScoreBins"] = df["CreditScore"].apply(custom_binning_credit_score)
del df["CreditScore"]
    
df.head(10)
```

```python
df.to_csv("../week--02/data/processed/CEHHbInToW_binning_handled_values.csv", index=False)
```
